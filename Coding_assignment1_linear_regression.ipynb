```python
# Creating data matrices for X and Y

X = df.drop('quality', axis=1)

y = df['quality']

X = X.to_numpy()
y = y.to_numpy()

y = y.reshape(-1)

X_shape, y_shape = X.shape, y.shape
X_shape, y_shape
```

```python
# splitting data into 80-20

np.random.seed(42) 

# we are shuffling the data to have randomness in the data
shuffled_indices = np.random.permutation(len(df))
shuffled_df = df.iloc[shuffled_indices]

X_shuffled = shuffled_df.drop('quality', axis=1).to_numpy()
y_shuffled = shuffled_df['quality'].to_numpy()

num_training_samples = int(len(df) * 0.80)

X_train = X_shuffled[:num_training_samples]
y_train = y_shuffled[:num_training_samples]
X_test = X_shuffled[num_training_samples:]
y_test = y_shuffled[num_training_samples:]

X_train_shape, X_test_shape, y_train_shape, y_test_shape = (
    X_train.shape, X_test.shape, y_train.shape, y_test.shape
)
X_train_shape, X_test_shape, y_train_shape, y_test_shape
```

```python
# Step 1: Add a Bias Term
X_train_bias = np.c_[np.ones(X_train.shape[0]), X_train]
X_test_bias = np.c_[np.ones(X_test.shape[0]), X_test]

# Step 2: Compute the Weights using the Normal Equation
w_closed_form = np.linalg.inv(X_train_bias.T @ X_train_bias) @ X_train_bias.T @ y_train

# Step 3: Make Predictions
y_train_pred = X_train_bias@w_closed_form
y_test_pred =X_test_bias@w_closed_form
```

```python
y_train_pred
```

```python
y_test_pred
```

```python


# Calculate Sum of Squared Errors
sse_train = np.sum((y_train_pred - y_train) ** 2)
sse_test = np.sum((y_test_pred - y_test) ** 2)



sse_train, sse_test
```

```python
# Plotting the predictions vs the actual data values for both train and test sets

# Setting up the plot
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
fig.suptitle('Predictions vs Actual Data Values (Ground Truth)')

# Plot for Training Data
axes[0].scatter(y_train, y_train_pred, alpha=0.6)
axes[0].plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'k--', lw=3)
axes[0].set_xlabel('Actual Values (Ground Truth)')
axes[0].set_ylabel('Predictions')
axes[0].set_title('Training Data')

# Plot for Testing Data
axes[1].scatter(y_test, y_test_pred, alpha=0.6, color='red')
axes[1].plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=3)
axes[1].set_xlabel('Actual Values (Ground Truth)')
axes[1].set_ylabel('Predictions')
axes[1].set_title('Testing Data')

plt.show()
```
